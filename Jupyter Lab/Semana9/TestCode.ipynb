```python
def sec_pares(ini=0, fin=100):
    if ini % 2 != 0:
        num = ini + 1
    
    while num < fin:
        yield num
        num += 2
```

```python
for pares in sec_pares(1, 20):
    print(pares)
```

```python
class Resistencia:
    def __init__(self, val=1000):
        self.val = val
        
    def __add__(self, other):
        return Resistencia(self.val + other.val)
    
    def __floordiv__(self, other):
        return 1 / (1/self.val + 1/other.val)
    
    def __repr__(self):
        return f"Resistencia(val={self.val} ohmios)"
```

```python
R1 = Resistencia(1000)
R2 = Resistencia(1000)

R1 // R2
```

```python
%%file gen_reporte.py
from collections import namedtuple
import csv
import sys
import os

if len(sys.argv) < 2:
    print("Genera un reporte a partir deun archivo CSV")
    print("   Uso: > python gen_reporte.py <archivo_csv>")
    sys.exit()
    
FILE = sys.argv[1]
TEXT = FILE[:-4] + ".txt"
Alumno = namedtuple("Alumno", ["nombre_completo", "notas", "nota_final"])
alumnos = []

# Abrir el archivo CSV y leer los campos
with open(FILE, encoding="utf-8") as file:
    reader = csv.reader(file)
    _ = next(reader)

    # Se genera la lista de alumnos
    for line in reader:
        arg1 = line[1] + ', ' + line[2]      
        arg2 = line[7:62:6]
        arg3 = sum([int(x) for x in arg2])
        alumnos.append(Alumno(arg1, arg2, arg3))
    
# Se ordenan los alumnos por nota
alumnos = sorted(alumnos, key=lambda x: x.nota_final, reverse=True)

# Se guarda el reporte
with open(TEXT, mode='w', encoding='utf-8') as file:
    file.write(f"{'NOMBRE COMPLETO':40}---- PREGUNTAS ----   NOTA\n")
    file.write("=" * 70)
    file.write("\n")
    
    for alumno in alumnos:
        file.write(f"{alumno.nombre_completo:40}")
        
        for nota in alumno.notas:
            file.write(f"{nota:2}")
        else:
            file.write(f"   {alumno.nota_final}\n")

print("Reporte generado:", os.path.abspath(TEXT))
```

```python
%%file tipo_cambio.py
import tkinter as tk
import requests

url = "https://deperu.com/api/rest/cotizaciondolar.json"
r = requests.get(url)
data = r.json()

tipo_compra = data['Cotizacion'][0]['Compra']
tipo_venta = data['Cotizacion'][0]['Venta']

class App:
    def __init__(self, master):
        self.master = master
        
        self.var_monto = tk.StringVar()
        
        frm = tk.Frame(self.master)
        frm.pack(padx=10, pady=10)
        
        frm1 = tk.Frame(frm)
        frm2 = tk.Frame(frm)
        frm3 = tk.Frame(frm)
        
        frm1.pack(padx=10, pady=10)
        frm2.pack(padx=10, pady=10)
        frm3.pack(padx=10, pady=10)
                
        self.lblMonto = tk.Label(frm1, text="Monto")
        self.entMonto = tk.Entry(frm1, textvariable=self.var_monto)
        self.btnCompra = tk.Button(frm2, text=f"Compra\n{tipo_compra}", width=12, command=self.compra)
        self.btnVenta = tk.Button(frm2, text=f"Venta\n{tipo_venta}", width=12, command=self.venta)
        self.lblTotal = tk.Label(frm3, text="")
        
        self.lblMonto.grid(row=0, column=0, padx=5, pady=5)
        self.entMonto.grid(row=0, column=1, padx=5, pady=5)
        self.btnCompra.grid(row=1, column=0, padx=5, pady=5)
        self.btnVenta.grid(row=1, column=1, padx=5, pady=5)
        self.lblTotal.grid(row=2, column=0, columnspan=2, padx=5, pady=5)
       
    
    def compra(self):
        self.lblTotal.config(text=f"$ {float(self.var_monto.get()) / float(tipo_compra):,.2f}")
    def venta(self):
        self.lblTotal.config(text=f"S/. {float(self.var_monto.get()) * float(tipo_venta):,.2f}")
    
root= tk.Tk()
app = App(root)
root.mainloop()

```
