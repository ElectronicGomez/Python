```python
# from tkinter import *
import tkinter as tk
import tkinter.ttk as ttk
import time
import csv
import sqlite3
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg, NavigationToolbar2Tk

regiones = ['ASIA (Ex, NEAR EAST)','EASTERN EUROPE','NORTHERN AFRICA',
                    'LATIN AMER &CARIB','WESTERN EUROPE']
        
paises=[ 'Colombia', 'Ecuador', 'Chile', 'Peru','Bolivia']

class Database:
    def __init__(self, master):
        self.master = master
        self.master.title("ESTADÍSTICAS MUNDIALES")
        self.master.geometry("1200x650+100+10")
        self.master.resizable(0,0)
        
        self.valor = IntVar()
        self.lista_poblacion=[]
        self.lista_migracion=[]
        self.lista_mortalidad=[]
        self.lista_alfabetismo=[]
        
        
        #Frames
        frm = tk.Frame(self.master)
        frm1 = Frame(frm)
        frm2 = Frame(frm)
        frm_regiones = LabelFrame(frm1, text="Regiones")
        frm_paises = LabelFrame(frm1, text="Paises")
        frm_grafico = LabelFrame(frm2, text="Data Graph")
        frm_estad = LabelFrame(frm2, text="Estadísticas")
        
        frm.pack()
        frm1.pack(side='left', padx=5, pady=5, anchor='n')
        frm2.pack(side='left', padx=5, pady=5, anchor='n')
        frm_regiones.pack(fill='both', padx=10)
        frm_paises.pack(fill='both', padx=10, pady=50, anchor='s')
        frm_grafico.pack()
        frm_estad.pack(side='left', expand=True, fill='x')
                             
        
        #Crear Widgets
        #scroll y listbo
        self.scrY = tk.Scrollbar(frm_regiones, orient='vertical')
        self.listReg = tk.Listbox(frm_regiones, height=8) 
        self.listReg.config(yscrollcommand=self.scrY.set)
        self.scrY.config(command=self.listReg.yview)                
        
        self.scrY2 = tk.Scrollbar(frm_paises, orient='vertical')
        self.listPais = tk.Listbox(frm_paises, height=8) 
        self.listPais.config(yscrollcommand=self.scrY2.set)
        self.scrY2.config(command=self.listPais.yview)
        
        #Radiobutton
        self.estad1 = Radiobutton(frm_estad, text="Población", variable=self.valor, value=1, state=DISABLED, command=self.select_estad)
        self.estad2 = Radiobutton(frm_estad, text="Migración", variable=self.valor, value=2, state=DISABLED, command=self.select_estad) 
        self.estad3 = Radiobutton(frm_estad, text="Mortalidad", variable=self.valor, value=3, state=DISABLED, command=self.select_estad)
        self.estad4 = Radiobutton(frm_estad, text="Alfabetismo", variable=self.valor, value=4, state=DISABLED, command=self.select_estad)
        self.estad1.select()
        
        #Colocar los WIDGETS
        
        self.listReg.pack(side=tk.LEFT)
        self.scrY.pack(side=tk.LEFT, expand=True, fill=tk.Y)
        self.listReg.bind("<<ListboxSelect>>", self.print_selected)

        self.listPais.pack(side=tk.LEFT)
        self.scrY2.pack(side=tk.LEFT, expand=True, fill=tk.Y)
        self.listPais.bind("<<ListboxSelect>>", self.print_selected2)
        
        self.estad1.grid(row=0, column=0, padx=70, pady=15, sticky='w')
        self.estad2.grid(row=0, column=1, padx=70, pady=15, sticky='w')
        self.estad3.grid(row=0, column=2, padx=70, pady=15, sticky='w')
        self.estad4.grid(row=0, column=3, padx=70, pady=15, sticky='w')
        
        self.listReg.bind('<Double-Button-1>', self.select_rad)
        self.listPais.bind('<Double-Button-1>', self.select_rad)
        
        for region in regiones:
            self.listReg.insert(tk.END, region)
        for pais in paises:
            self.listPais.insert(tk.END, pais)
        #Figure
        self.fig, self.ax = plt.subplots(figsize=(14, 6))
        self.fig.set_facecolor("#F0F0F0")
        self.grafica = FigureCanvasTkAgg(self.fig, master=frm_grafico)
        self.grafica.get_tk_widget().pack(side='top')
        
        self.herramientas = NavigationToolbar2Tk(self.grafica, frm_grafico)
        self.herramientas.update()
        
        self.ax.set_title("Country Data",{"color":"blue"}, fontweight="bold")
        
        
    def print_selected(self, event):
        print(self.listReg.get(self.listReg.curselection()))
    def print_selected2(self, event):
        print(self.listPais.get(self.listPais.curselection()))
    
    def select_rad(self, event):
        self.estad1.config(state='normal')
        self.estad2.config(state='normal')
        self.estad3.config(state='normal')
        self.estad4.config(state='normal')
    
    def select_estad(self):
        if self.valor.get()==1:
            self.ax.cla()
            pass
                
        elif self.valor.get()==2:
            self.ax.cla()
            pass
        
        elif self.valor.get()==3:
            self.ax.cla() 
            pass
               
        elif self.valor.get()==4:
            self.ax.cla() 
            pass
            
        
root = tk.Tk()
app = Database(root)
root.mainloop()
```
